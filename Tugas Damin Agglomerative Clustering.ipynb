```python
# Do the neccessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from sklearn.datasets import make_blobs
```

```python
# Create a blob of 200 data points
# dataset = make_blobs(n_samples = 200,
#                     n_features = 2,
#                     centers = 4,
#                     cluster_std = 1.6,
#                     random_state = 50)

# Import Dataset dari file .csv
importdata = pd.read_csv('datasettugasdamin.csv')

col1 = importdata['x']
col2 = importdata['y']

df = pd.DataFrame ({
    'x' : col1,
    'y' : col2
})
```

```python
# points = dataset[0]
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# Create a dendrogram
dendrogram = sch.dendrogram(sch.linkage(df,method='ward'))
```

```python
plt.scatter(df['x'],df['y'])
```

```python
# Perform the actual clustering
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage='ward')
```

```python
y_hc = hc.fit_predict(df)
```

```python
print(y_hc)
```

```python
plt.scatter(df['x'],df['y'], c=y_hc, cmap = 'rainbow')
# plt.scatter(points[y_hc ==1,0], points[y_hc == 1,1], s = 100, c = 'yellow')
# plt.scatter(points[y_hc ==2,0], points[y_hc == 2,1], s = 100, c = 'red')
# plt.scatter(points[y_hc ==3,0], points[y_hc == 3,1], s = 100, c = 'green')
```
